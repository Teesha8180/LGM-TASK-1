### Importing required Libraries

```python
import cv2   # We will use OpenCV to read image
import matplotlib.pyplot as plt   # For visualization
from PIL import Image    # We will use Image from PIL to convert array to image
from IPython.display import display   # To display the image
```

### Reading Image

```python
image = cv2.imread("jellyfish.jpg")
```

### Showing image using IPython Library

```python
display(Image.fromarray(image[:,:,::-1]))   # operation on array of image to convert BGR image to RGB image
                                            # As display uses BGR color scheme
```

### Converting Original Image to Grayscale Image 

### For this we will use cvtColor function of OpenCV

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

### Inverting Grayscale Image

### For this we will use bitwise_not function of OpenCV

```python
invert_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(invert_image))
```

### Blur Image

### For this we will use Gaussian Function in OpenCV

```python
blur_image = cv2.GaussianBlur(invert_image, (111,111), 0)
display(Image.fromarray(blur_image))
```

### Inverted Blur Image

### After this we can create the pencil sketch by mixing the GrayScale Image with the inverted blurry Image

```python
invert_blur_image = cv2.bitwise_not(blur_image)
display(Image.fromarray(invert_blur_image))
```

### Creating the Pencil sketch

### This can be done by performing bit-wise division between the grayscale image and the inverted-blurred image.

```python
sketch_image = cv2.divide(gray_image, invert_blur_image, scale=(256.0))
display(Image.fromarray(sketch_image))
```
